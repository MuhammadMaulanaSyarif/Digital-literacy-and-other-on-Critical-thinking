## Defined X and Y

```python
x = df.drop(['Kemampuan Berpikir Kritis'],axis=1).values
y = df['Kemampuan Berpikir Kritis'].values
```

```python
print(x)
```

```python
print(y)
```

## Train Data Set X and Y

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split (x,y,test_size=0.33,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg_1 = LinearRegression()
reg_1.fit(x_train,y_train)
```

```python
y_pred = reg_1.predict(x_test)
print(y_pred)
```

## Model regression Evaluation

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

## Show Regression Diagaram

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test,y_pred)
plt.xlabel('Nilai Aktual')
plt.ylabel('Nilai Prediksi')
plt.title('Prediksi Kemampuan Berpikir Kritis Terhadap Literasi Digital, Motivasi Belajar, dan Kemandirian Belajar')
```

## Difference between actual and predicted values

```python
pred_y_df = pd.DataFrame({'Nilai Aktual':y_test,'Nilai Prediksi':y_pred,'selisih':y_pred-y_test})
pred_y_df
```
